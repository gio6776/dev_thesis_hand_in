## Final Tranformation of the Output
1 -  Extract Input from user_question columns

```python
# read excel file to df
df = pd.read_excel(r'G:\My Drive\Profissional & Acadêmico\Mestrados\DTU\5_thesis\dev_thesis\data\de_data\processed_results\langsmith\langsmith_results_clean_non-technical-stakeholders.xlsx')
```

```python
import pandas as pd
import ast

# Assuming you have a dataframe 'df' with a column 'user_question'
def extract_input(s):
    try:
        # Use ast.literal_eval to safely evaluate the string as a Python literal
        data = ast.literal_eval(s)
        # Extract the 'input' from the first dictionary in the list
        return data[0]['input']
    except (ValueError, SyntaxError, KeyError, IndexError):
        # Return None or some error indicator if parsing fails
        return None

# Apply the function to create a new column
df['extracted_input'] = df['user_question'].apply(extract_input)

# Now 'extracted_input' column will contain the input string
print(df['extracted_input'])

df
```
