```python
import json

def parse_txt_to_json(filepath):
    # Initialize variables
    questions = []
    current_complexity = None

    # Open the file and process each line
    with open(filepath, 'r') as file:
        for line in file:
            line = line.strip()  # Remove leading/trailing whitespaces
            if not line:
                continue  # Skip empty lines
            
            # Check for complexity level
            if line in ["SIMPLE", "MODERATE", "COMPLEX"]:
                current_complexity = line
                continue


            # Assuming the line is a question, construct the dictionary
            question_dict = {
                "table_id": "amazon_orders",
                "question": line,
                "complexity": current_complexity,
                "query": None,
                "result": None
            }
            questions.append(question_dict)
    
    # Convert the list to a JSON-formatted string
    return json.dumps(questions, indent=4)

# Example usage
filepath = r"G:\My Drive\Profissional & Acadêmico\Mestrados\DTU\5_thesis\dev_thesis\data\de_data\questions_eval\questions_answerable.txt"  # Replace with your actual file path
json_output = parse_txt_to_json(filepath)
print(json_output)
```
