## Save all queries from a Dashboard

```python
response_json = mb.get('/api/dashboard/393')
response_json
```

```python
import json

# Processing the response to match the desired structure
dashboard_data = {
    "dashboard_id": 2,
    "dashboard_name": "Bol Sales",
    "cards": []
}

dashcards = response_json['dashcards']

for dashcard in dashcards:
    if 'name' in dashcard['card']:
        # Simulating the API call to get query results
        query_result = mb.post('/api/card/' + str(dashcard['card']['id']) + '/query')['data']['cols']

        # Processing the output schema from the query result
        output_schema_parts = []
        for field in query_result:
            # Concatenating "name" and "base_type" into a single string
            field_str = f"{field['name']} - {field['base_type']}"
            output_schema_parts.append(field_str)
        
        # Joining all parts with a comma to create a single string
        output_schema_str = ", ".join(output_schema_parts)
        
        card_info = {
            "id": dashcard['card']['id'],
            "name": dashcard['card']['name'],
            "query_type": dashcard['card']['query_type'],
            "query": dashcard['card']['dataset_query']['native']['query'],
            "output_schema": output_schema_str
        }
        dashboard_data["cards"].append(card_info)

# Saving the structured data to a JSON file
with open('dashboard_data.json', 'w') as json_file:
    json.dump(dashboard_data, json_file, indent=4)

print("Data saved to 'dashboard_data.json'.")
```
